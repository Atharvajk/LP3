```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient_descent(starting_x, learning_rate, iterations):
    x = starting_x
    x_history = [x]  # To store the x values over iterations for visualization
    for i in range(iterations):
        gradient = derivative(x)
        x = x - learning_rate * gradient
        x_history.append(x)
    
    return x, x_history
```

```python
# Parameters
starting_x = 2
learning_rate = 0.1 # aplha
iterations = 20
```

```python
local_minimum, x_history = gradient_descent(starting_x, learning_rate, iterations)
print("Local Minima is ",local_minimum)
```

```python
# Plotting the function and gradient descent steps
x_vals = np.linspace(-6, 6, 100)  # x values for plotting
y_vals = (x_vals + 3)**2  # y = (x + 3)^2

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="y = (x + 3)^2", color="blue")
plt.scatter(x_history, [(x + 3)**2 for x in x_history], color="red", s=40, label="Gradient Descent Steps")
plt.plot(x_history, [(x + 3)**2 for x in x_history], color="red", linestyle="--", alpha=0.5)

# # Mark starting and ending points
plt.scatter([starting_x], [(starting_x + 3)**2], color="green", s=100, label="Starting Point (x=2)")
plt.scatter([local_minimum], [(local_minimum + 3)**2], color="purple", s=100, label="Local Minimum")

# Labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent on y = (x + 3)^2")
plt.legend()
plt.grid(True)
plt.show()
```

1. What is Gradient Descent?
Gradient Descent is an optimization algorithm used to minimize (or maximize) a function by iteratively moving towards the direction of the steepest descent, which is defined by the negative of the gradient.

In simple terms, gradient descent helps us find the lowest point (minimum) on a function's curve. It’s widely used in machine learning and neural networks to minimize the cost function or loss.

1. Local Minimum and Local Maximum
In mathematical terms, a local minimum of a function is a point where the function value is lower than at any nearby points. Similarly, a local maximum is a point where the function value is higher than at any nearby points.

To visualize:

Imagine a mountainous landscape with hills and valleys.
The local minimum is the bottom of a valley—it’s the lowest point in the immediate area.
The local maximum is the peak of a hill—it’s the highest point in the immediate area.

2. Global Minimum and Global Maximum
A global minimum or global maximum is the lowest or highest point of the entire function.

Global Minimum: The absolute lowest point across the entire range of the function.
Global Maximum: The absolute highest point across the entire range of the function.
If we consider the entire landscape, the global minimum would be the lowest valley, and the global maximum would be the highest peak in the entire region.
